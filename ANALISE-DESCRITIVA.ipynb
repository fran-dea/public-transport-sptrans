```python
import pandas as pd
```

```python
rotas=pd.read_parquet("dados/rotaaas_historico.parquet")
viagens=pd.read_parquet("dados/viagens_historico.parquet")
paradas=pd.read_parquet("dados/paradas.parquet")
estacoes_metro_e_trem=pd.read_parquet("dados/estacoes_metro_e_trem.parquet")
```

```python
# Função para descrever cada DataFrame
def analise_df(df, nome):
    print(f"\n{'='*30} {nome.upper()} {'='*30}")
    print(f"Formato: {df.shape[0]} linhas x {df.shape[1]} colunas")
    print("\nTipos de dados:")
    print(df.info())
    

# Aplicar para cada DataFrame
analise_df(rotas, "rotas")
analise_df(viagens, "viagens")
analise_df(estacoes_metro_e_trem, "estacoes_metro_e_trem")
analise_df(paradas, "paradas")
```
